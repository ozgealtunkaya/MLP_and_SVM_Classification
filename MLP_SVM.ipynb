```python
import pandas as pd 
data=pd.read_csv('segment.csv', 
                names=['region-centroid-col','region-centroid-row','region-pixel-count','short-line-density-5','short-line-density-2','vedge-mean','vegde-sd','hedge-mean','hedge-sd','intensity-mean','rawred-mean','rawblue-mean','rawgreen-mean','exred-mean','exblue-mean','exgreen-mean','value-mean','saturatoin-mean','hue-mean','Label'])
classLabel=data.iloc[:, -1:].values 
print(classLabel)
data.head()
```

```python
from sklearn.model_selection import train_test_split  #veri kümesini eğitim ve test olarak ayırma kısmı
x_train,x_test,y_train,y_test=train_test_split(data.iloc[:,:-1],classLabel,test_size=0.20,random_state=0) #%20 test %80 eğitim
print(x_train)
```

```python
from sklearn.preprocessing import StandardScaler  
scaler=StandardScaler() 
scaler.fit(x_train) 
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier 
mlp=MLPClassifier(hidden_layer_sizes=(19,19,19), max_iter=500)
mlp.fit(x_train,y_train)
```

```python
predictions = mlp.predict(x_test) 
from sklearn.metrics import classification_report,confusion_matrix  
print(confusion_matrix(y_test,predictions)) 

print(classification_report(y_test,predictions))

from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

```python
import keras 
import tensorflow
from keras.models import Sequential
from keras.layers import Dense
classifier = Sequential()
classifier.add(Dense(units = 128, kernel_initializer = 'uniform', activation = 'relu', input_dim = 19,)) 
classifier.add(Dense(units = 64, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy'])
history = classifier.fit(x_train, y_train, validation_split=0.20, batch_size = 10, epochs = 5)

#Accuracy
loss, accuracy = classifier.evaluate(x_train, y_train, verbose=True)
print("Training Accuracy: {:.4f}".format(accuracy))
loss, accuracy = classifier.evaluate(x_test, y_test, verbose=False)
print("Testing Accuracy:  {:.4f}".format(accuracy))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

svc=SVC(kernel='poly')
svc.fit(x_train,y_train)
sonuc = svc.predict(x_test)
support = confusion_matrix(y_test,sonuc)
Accuracy = accuracy_score(y_test,sonuc)
print('Doğruluk / SVM poly : ' ,Accuracy)
print(support)
print(classification_report(y_test,sonuc))

print("","#######################################################################")

svc = SVC(kernel = 'linear')
svc = svc.fit(x_train,y_train) 
sonuc = svc.predict(x_test)
support = confusion_matrix(y_test,sonuc)
Accuracy =accuracy_score(y_test,sonuc)
print('Doğruluk / SVM linear : ' , Accuracy)
print(support)
print(classification_report(y_test,sonuc))

print("","#######################################################################")

svc = SVC(kernel = 'sigmoid')
svc.fit(x_train,y_train) 
sonuc = svc.predict(x_test)
support = confusion_matrix(y_test,sonuc)
Accuracy =accuracy_score(y_test,sonuc)
print('Doğruluk / SVM sigmoid : ' ,Accuracy)
print(support)
print(classification_report(y_test,sonuc))

print("","#######################################################################")

svc = SVC(kernel = 'rbf')
svc.fit(x_train,y_train) 
sonuc = svc.predict(x_test)
support = confusion_matrix(y_test,sonuc)
Accuracy =accuracy_score(y_test,sonuc)
print('Doğruluk / SVM rbf : ' ,Accuracy)
print(support)
print(classification_report(y_test,sonuc))
```
